```python
import matplotlib.pyplot as plt
import pandas as pd

# Your data
data = {
    'Year': [2018, 2019, 2020, 2021, 2022,2023],
    'BEVs': [52394,69368,119914,242209,473567,749686],
    'PHEVs': [88284,124692,158462,229701,339890,433157]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Calculate BEVs percentage
df['BEV Percentage'] = (df['BEVs'] / (df['BEVs'] + df['PHEVs'])) * 100

# Create a vertical stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))

bars = ax.bar(df['Year'], df['BEVs'], label='BEVs', color='blue')
ax.bar(df['Year'], df['PHEVs'], label='PHEVs', bottom=df['BEVs'], color='lightblue')

# Set labels and title with larger font size
ax.set_xlabel('Year (Q1)', fontsize=16)
ax.set_ylabel('Total EVs (in thousands)', fontsize=16)
ax.set_title('Total EVs in UK from 2018 to 2023', fontsize=18)
ax.legend(fontsize=18)

# Add BEV percentage labels just above the first bar with larger font size
for index, row in df.iterrows():
    bev_percentage_label = f'{row["BEV Percentage"]:.2f}%'
    ax.text(row['Year'], row['BEVs'] + 50, bev_percentage_label,
            ha='center', va='bottom', color='blue', fontsize=14)

# Set y-axis to show in thousands
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x/1e3))))
ax.tick_params(axis='both', which='major', labelsize=14)
# Show the plot
plt.show()
```
